## Data loading

The dataframe `df_meta` contains demographic information about the patients, as well as diagnoses either given by the cardiologist or automatically generated by the machine that measured the ECG. To each ECG in `df_meta` corresponds a `.csv` file in the `ecg/` folder, indicated by the `ecg_file_path` variable in the dataframe. Each ECG contains 12 columns (one for each derivation), and 5000 rows. Each row corresponds to the electric signal at a given time for the 12 electrodes. The sampling rate is 500Hz, meaning that each ECG corresponds to a 10-second measurement.

```python
df_meta = pd.read_pickle("df_meta.pkl")
```

```python
ecg_example = pd.read_csv(df_meta.iloc[0].ecg_file_path)
```
